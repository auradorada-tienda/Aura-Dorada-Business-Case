## **ETAPA 3 — DATA PREPARATION**

Limpieza y preparación de los datos.

**3.1 Modelado de datos**

```python
# Carga del archivo Excel y separación de las hojas en diferentes tablas (modelo de datos por dimensiones)

xls = pd.ExcelFile('/content/4 - Dataset_original _ Aura_Dorada.xlsx')

vendas = pd.read_excel(xls, sheet_name='VENDAS')
produtos = pd.read_excel(xls, sheet_name='PRODUTOS')
lojas = pd.read_excel(xls, sheet_name='LOJAS')
campanhas = pd.read_excel(xls, sheet_name='CAMPANHAS')
```

El dataset está estructurado en un modelo de datos compuesto por una tabla de hechos (ventas) y varias tablas dimensionales (productos, tiendas y campañas).

Esta organización permite analizar el comportamiento de las ventas desde diferentes perspectivas del negocio, como producto, ubicación y campañas comerciales, facilitando un análisis más completo y segmentado.

**3.2 Corrección de inconsistencias**

**3.2.1 - Normalización de campañas**

```python
# Normalización de la variable campaña
ventas['campanha'] = ventas['campanha'].astype(str).str.replace('camp', '', regex=False)
ventas['campanha'] = ventas['campanha'].astype(int)
```

Se estandarizó la variable campaña eliminando el prefijo “camp” y convirtiendo los valores a formato numérico, con el objetivo de unificar la codificación y evitar duplicidades en el análisis.

**3.2.2 - Estandarización de tiendas**

```python
ventas['loja'] = ventas['loja'].replace({
    'BCN Centro': 'Barcelona Centro'
})
ventas[ventas['loja'].str.contains('BCN', case=False)]
```

Se estandarizó la variable tienda unificando la nomenclatura de “BCN Centro” a “Barcelona Centro”, con el objetivo de evitar duplicidades y garantizar la consistencia en el análisis por ubicación.

Esta corrección permite una agregación más precisa de las ventas por tienda.

**3.2.3 - Normalización de productos**

```python
ventas['produto'] = ventas['produto'].str.strip().str.title()
```

Se estandarizó la variable producto eliminando espacios en blanco y aplicando formato tipo título (Title Case), con el objetivo de unificar la escritura de los nombres y mejorar la consistencia en el análisis.

Esta transformación ayuda a reducir variaciones en la escritura de los productos y facilita su agregación correcta.

**3.3 Calidad de datos (nulos + duplicados)**

```python
# Verificación de valores nulos
ventas.isnull().sum()
productos.isnull().sum()
tiendas.isnull().sum()
campañas.isnull().sum()

# Verificación de duplicados
ventas.duplicated().sum()
productos.duplicated().sum()
tiendas.duplicated().sum()
campañas.duplicated().sum()
```

Se realizó un análisis de calidad de datos mediante la verificación de valores nulos y registros duplicados en todas las tablas del dataset.

Los resultados muestran que no existen valores nulos ni duplicados, lo que indica una buena calidad de los datos y permite continuar con las siguientes fases del análisis sin necesidad de tratamientos adicionales.

## Etapa 5 - Visualization
Creación de gráficos y dashboards.

El objetivo de esta etapa es representar visualmente los principales patrones identificados en el análisis, facilitando la interpretación de los datos y apoyando la toma de decisiones estratégicas.

### **5.1 Ventas por producto**

```python
import matplotlib.pyplot as plt

ventas.groupby('produto')['vendas_totais'].sum().sort_values().plot(
    kind='barh',
    figsize=(8,5)
)

plt.title('Ventas Totales por Producto')
plt.xlabel('Ventas Totales')
plt.ylabel('Producto')
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.show()
```

El gráfico confirma una fuerte concentración de ingresos en Aura Pods, que domina claramente frente al resto de productos, evidenciando una estructura de ventas altamente dependiente de este producto.

El resto del catálogo presenta una contribución significativamente menor, lo que sugiere una distribución desigual del rendimiento comercial.

### **5.2 Ventas por tienda**

```python
ventas.groupby('loja')['vendas_totais'].sum().sort_values().plot(
    kind='bar',
    figsize=(9,5)
)

plt.title('Ventas Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ventas Totales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

El gráfico muestra una distribución relativamente equilibrada de las ventas entre las diferentes tiendas, con valores bastante similares en la mayoría de ubicaciones. Sin embargo, se observa un rendimiento ligeramente superior en tiendas como Gracia, Barcelona Centro y Diagonal Mar, lo que podría indicar diferencias moderadas en el flujo de clientes o en el desempeño comercial de determinados puntos de venta.

### **5.3 Ventas por campaña**

```python
ventas.groupby('campanha')['vendas_totais'].sum().sort_values().plot(
    kind='bar',
    figsize=(7,5)
)

plt.title('Impacto de las Campañas en Ventas')
plt.xlabel('Campaña')
plt.ylabel('Ventas Totales')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

La visualización muestra diferencias significativas en el rendimiento de las campañas, destacando la campaña 0 como la principal generadora de ventas durante el período analizado.

Este resultado sugiere que determinadas estrategias promocionales pueden tener un impacto considerable en el comportamiento de compra y en el volumen total de ventas.

### **5.4 Evolución mensual de ventas**

```python
ventas.groupby('mes')['vendas_totais'].sum().sort_index().plot(
    marker='o',
    figsize=(8,5)
)

plt.title('Evolución Mensual de Ventas')
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

La evolución mensual de las ventas muestra un comportamiento generalmente estable a lo largo del período analizado, con valores relativamente homogéneos entre los distintos meses.

Sin embargo, se identifica un pico claro en el mes 4 (40.453,25), que se sitúa por encima del resto de meses del año. Este incremento sugiere la posible influencia de factores puntuales, como acciones promocionales o variaciones en la demanda.

A pesar de esta variación, no se observan fluctuaciones extremas en el resto del período, lo que indica una demanda relativamente constante en el tiempo.

**5.5 Relación entre productos y tiendas**

```python
import seaborn as sns

pivot = ventas.pivot_table(
    values='vendas_totais',
    index='loja',
    columns='produto',
    aggfunc='sum'
)

plt.figure(figsize=(10,6))

sns.heatmap(
    pivot,
    annot=True,
    fmt='.0f',
    cmap='Blues'
)

plt.title('Relación entre Tiendas y Productos')

plt.show()
```

El mapa de calor permite analizar la distribución de las ventas entre productos y tiendas de forma comparativa, facilitando la identificación de patrones de consumo por ubicación. Se observa que Aura Pods concentra los valores más altos en prácticamente todas las tiendas, destacando especialmente en Gracia, Diagonal Mar y El Born.


Por otro lado, productos como Aura Device presentan niveles de ventas más bajos y homogéneos entre ubicaciones, lo que sugiere un rendimiento más estable pero menos relevante en términos de ingresos. En conjunto, el análisis evidencia que el rendimiento de los productos no es uniforme entre tiendas, sino que existe una clara concentración en productos específicos, principalmente Aura Pods.

Estas oportunidades pueden servir como base para impulsar el crecimiento de las ventas y contribuir al objetivo de incremento del volumen total de ventas planteado en este proyecto.



### **6.3 Limitaciones del análisis**

El análisis presenta algunas limitaciones que deben ser consideradas para una correcta interpretación de los resultados:

*   El estudio se basa exclusivamente en datos internos de ventas, sin incluir variables externas como inversión en marketing, tráfico de clientes o contexto económico, lo que limita la explicación de algunos comportamientos observados.

*   No se dispone de información cualitativa sobre el comportamiento del consumidor, lo que dificulta profundizar en las causas detrás de las variaciones de ventas entre productos y campañas.

*   El análisis se centra en un período temporal específico, por lo que los resultados pueden no reflejar tendencias a largo plazo.

*   Aunque los datos han sido limpiados y estandarizados durante la fase de preparación, siempre existe una limitación inherente a cualquier proceso de análisis basado en datos históricos.
